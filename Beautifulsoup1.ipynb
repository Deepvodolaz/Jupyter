```python
import requests
from bs4 import BeautifulSoup

url = "https://vkusnyesushi.ru"
response = requests.get(url)
response
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content, 'lxml')
soup
```

```python
print(soup.prettify())
```

```python
lst = soup.find_all('h5')
lst
```

```python
def clean_item(my_item):
    position = my_item.find('<a')
    return my_item[4:position]

print(clean_item('<h5>Text, labels and annotations<a clas'))
```

```python
for item in lst:
    print(clean_item(str(item)))
```

```python
def clean_item2(my_item):
    position = my_item.find('<i')
    return my_item[4:position]

for item in lst:
    print(clean_item2(str(item)))
```
